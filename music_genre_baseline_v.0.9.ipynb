##### Получение эмбединга.

````python
def my_get_embedding(img: Image, model: Any) -> np.array:
    """
    Takes an image and a model as input and returns the image embedding.

    Args:
        img (PIL.Image): Input image.
        model (torch.nn.Sequential): ResNet50 model or a similar model.

    Returns:
        np.ndarray: Image embedding as a NumPy array.
    """

    # Image preprocessing
    preprocess = transforms.Compose([
        transforms.Resize((224, 224)), # Resized with proportion saved, missing pixels added.
        #transforms.Resize(256), # Simple resize. No propotion saved.
        #transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    
    img_tensor: torch.Tensor = preprocess(img)

    # Check CUDA availability
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    img_tensor = img_tensor.to(device)

    img_tensor = img_tensor.unsqueeze(0)  # Add batch dimension

    # Forward pass through the network to extract embeddings
    with torch.no_grad():
        features: torch.Tensor = model(img_tensor)

    # Return torch.Tensor as np.array
    embedding: np.ndarray = features.detach().reshape(-1).cpu().numpy()
    return embedding
````

##### Создание датафрейма данных.

````python
def my_create_music_album_df(music_folder_path: str) -> pd.DataFrame:
    """
    Creates a DataFrame based on the folder structure of music albums.

    Args:
        music_folder_path (str): Path to the folder containing music albums.

    Returns:
        pd.DataFrame: DataFrame with information about music albums.
    """

    data: list[dict] = []

    # Iterate over each folder inside music_folder_path
    for genre_folder in os.listdir(music_folder_path):
        genre_folder_path: str = os.path.join(music_folder_path, genre_folder)
        
        # Check if the element is a folder
        if os.path.isdir(genre_folder_path):
            for album_file in os.listdir(genre_folder_path):
                index: str = f'{genre_folder}:{album_file}'
                full_file_name: str = os.path.join(genre_folder_path, album_file)
                data.append({'genre': genre_folder, 
                             'file_name': album_file, 
                             'index': index, 
                             'full_file_name': full_file_name})

    # Create a DataFrame from the list
    df: pd.DataFrame = pd.DataFrame(data)
    return df
````

##### Генерация эмбедингов.

````python
def my_compute_embeddings(df: pd.DataFrame, model: Any) -> tuple[np.ndarray, np.ndarray]:
    """
    Calculates image embeddings from the DataFrame using the model and writes them into a NumPy array.

    Args:
        df (pd.DataFrame): DataFrame with information about the images.
        model (torch.nn.Sequential): Model for calculating the embeddings.

    Returns:
        tuple[np.ndarray, np.ndarray]: NumPy array with the image embeddings and NumPy array with the indices.
    """
    
    embeddings: list[np.array] = []
    indices: list[str] = []
    df_length = len(df)

    # Iterate over each row in the dataframe
    for i, row in df.iterrows():
        img: Image = Image.open(row['full_file_name']).convert('RGB')
        embedding: np.array = my_get_embedding(img, model)
        embeddings.append(embedding)
        indices.append(row['index'])
        print(f'Percentage of task completion: {i/df_length:.0%}', end='\r')

    # Convert the list of embeddings and indexes into numpy arrays
    embeddings_array: np.array = np.array(embeddings)
    indices_array: np.array = np.array(indices)

    return embeddings_array, indices_array
````

````python
def my_display_images_from_dataframe(dataframe: pd.DataFrame, images_per_row: int = 3, num_rows: int = 3) -> None:
    """
    Display images from a dataframe as a grid.

    Args:
        dataframe (pd.DataFrame): DataFrame with information about the images.
        images_per_row (int, optional): Number of images per row in the grid. Defaults to 5.
        num_rows (int, optional): Number of rows in the grid. Defaults to 3.
    """
    
    # Create a new figure with the specified size
    fig: plt.Figure = plt.figure(figsize=(10, 10))

    # Iterate over each image in the dataframe, limiting the total number of images to display
    for i, image_filename in enumerate(dataframe['full_file_name'][:images_per_row * num_rows], start=1):

        if not os.path.exists(image_filename):
            print(f'File not found: {image_filename}')
            continue

        genre: str = dataframe[dataframe['full_file_name'] == image_filename]['genre'].values[0]
        img: Image.Image = Image.open(image_filename)

        ax: plt.Axes = fig.add_subplot(num_rows, images_per_row, i)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(f'Image #{i}\nGenre: {genre}', fontsize=TITLE_FONTSIZE)

    # Adjust the layout of the subplots to fit nicely
    plt.tight_layout()
    
    # Show the plot
    plt.show()
    
    return
````

##### Получение информации об изображениях.

````python
def my_extract_image_info(dataframe: pd.DataFrame) -> pd.DataFrame:
    """
    Extracts information about images from the DataFrame.

    Parameters:
        dataframe (pd.DataFrame): DataFrame containing information about the images.

    Returns:
        pd.DataFrame: DataFrame with information about image sizes.
    """
    
    # Create an empty list to store information about image sizes
    image_info_list: list[dict] = []

    # Iterate over each row in the input dataframe
    for index, row in dataframe.iterrows():
        # Get the full path to the image
        image_path: str = row['full_file_name']
        
        # Check if the file exists
        if os.path.exists(image_path):
            # Get the file size
            file_size: int = os.path.getsize(image_path)
            
            try:
                with Image.open(image_path) as img:
                    width: int
                    height: int
                    width, height = img.size
            except Exception as e:
                print(f"Error processing image {row['full_file_name']}: {e}")
                width, height = 0, 0
            
            # Add the information to the list
            image_info_list.append({'full_file_name': row['full_file_name'],
                                    'width': width,
                                    'height': height,
                                    'file_size': file_size})
        else:
            print(f'File not found: {image_path}')

    # Create a new dataframe from the list
    image_info_df = pd.DataFrame(image_info_list)
    return image_info_df
````

##### Поиск дубликатов по эмбедингам.

````python
def my_find_duplicates(arr: np.ndarray, threshold: float = 0.0) -> np.ndarray:
    """
    Finds duplicate vectors in the array.

    Parameters:
        arr (np.ndarray): Input array with vectors.
        threshold (float): Threshold value for the distance between vectors,
                           below which they will be considered duplicates.
                           Default is 0.0.

    Returns:
        np.ndarray: Array of indices of duplicates in the original array.
    """

    # Initialize the NearestNeighbors model
    nn_model: NearestNeighbors = NearestNeighbors(n_neighbors=3, algorithm='auto', metric='euclidean')
    nn_model.fit(arr)
    
    # Find the indices and distances to the nearest neighbors for each vector
    distances: np.ndarray
    indices: np.ndarray
    distances, indices = nn_model.kneighbors(arr)
    
    # Filter out pairs of neighboring vectors where the distance between them is less than the threshold value
    duplicates_mask: np.ndarray = distances[:, 1] <= threshold
    duplicates_indices: np.ndarray = np.where(duplicates_mask)[0]
    
    return duplicates_indices
````

````python
def my_plot_confusion_matrix_v2(true_labels: list[str],
                                predicted_labels: list[str],
                                labels: list[str],
                                title: str,
                                ax=None) -> None:
    """
    Plot the confusion matrix.

    Args:
        true_labels (List[str]): List of true labels.
        predicted_labels (List[str]): List of predicted labels.
        labels (List[str]): List of all classes (labels).
        title (str): Title for the confusion matrix plot.
        ax: Axes object for subplot. If None, a new subplot will be created.
    """
    
    # Compute the normalized confusion matrix
    cm: np.ndarray = confusion_matrix(true_labels, predicted_labels, labels=labels)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    cm = np.round(cm, 1)
    
    # Create ConfusionMatrixDisplay object
    display: Any = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    
    # Plot the confusion matrix on the specified Axes or a new subplot
    if ax is None:
        plt.figure(figsize=(6, 5))
        ax = plt.gca()
    
    display.plot(ax=ax, cmap='Blues', xticks_rotation='vertical')
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title(title, pad=20)
````

#### Классы.

##### Класс MyFaissClassificator.

````python
class MyFaissClassificator():
    """
    Class for using FAISS to predict labels based on nearest neighbors.
    
    Attributes:
        dim (int): Dimensionality of vectors.
        index (faiss.index): FAISS index for nearest neighbor search.
        k (int): Number of nearest neighbors to search.
        train_labels (list[str]): List of labels for the training data.
    """
    
    def __init__(self, dim: int, k: int) -> None:
        """
        Initialize the MyFaissPredictor class.
        
        Parameters:
            dim (int): Dimensionality of vectors.
            k (int): Number of nearest neighbors to search.
        """
        
        self.dim: int = dim
        self.index: Any = faiss.IndexFlatL2(dim)
        self.k: int = k
        self.index.seed: int = RANDOM_STATE 

                    
    def fit(self, embeddings_train: np.ndarray, train_labels: list[str]) -> None:
        """
        Train the model on the training embeddings.
        
        Parameters:
            embeddings_train (np.array): Array of training embeddings.
            train_labels (list[str]): List of labels for the training data.
        """
        
        self.train_labels = train_labels
        self.index.add(embeddings_train)
        
    def predict(self, embeddings_test: np.ndarray) -> list[str]:
        """
        Predict labels for the test embeddings based on nearest neighbors.
        
        Parameters:
            embeddings_test (np.array): Array of test embeddings.
            
        Returns:
            list[str]: List of predicted labels for the test data.
        """
        
        distances: float
        indices: int
        distances, indices = self.index.search(embeddings_test, self.k)
        
        predicted_labels: List = []
        for neighbors in indices:
            neighbor_labels = [self.train_labels[i] for i in neighbors]
            predicted_label = max(set(neighbor_labels), key=neighbor_labels.count)
            predicted_labels.append(predicted_label)
            
        return predicted_labels
    
    def __call__(self, embeddings_test: np.ndarray) -> list[str]:
        """
        Call method to predict labels for the test embeddings.
        
        Parameters:
            embeddings_test (np.array): Array of test embeddings.
            
        Returns:
            list[str]: List of predicted labels for the test data.
        """
        
        return self.predict(embeddings_test)
````

##### Класс MyGridSearch.

````python
class MyGridSearch():
    """
    Class for performing Grid Search with MyFaissClassifier.
    
    Attributes:
        model_class (Any): The model class.
        param_grid (Dict[str, List[Any]]): The parameter grid for iteration.
        results (defaultdict): Dictionary to store search results.
        best_accuracy_score (float): Best accuracy score.
        best_f1_score (float): Best F1 score.
        best_params (Dict): Best parameters for the model.
        best_model (Any): Best model.
    """

    def __init__(self, model_class: Any, param_grid: Dict[str, List[Any]]) -> None:
        """
        Initialize the GridSearch object.

        Args:
            model_class (Any): The model class for which the search will be conducted.
            param_grid (Dict[str, List[Any]]): The parameter grid for iteration.

        """
        
        self.model_class: Any = model_class
        self.param_grid: Dict[str, List[Any]] = param_grid
        self.results: defaultdict = defaultdict(list)
        self.best_accuracy_score_: float = 0.0 
        self.best_f1_score_: float = 0.0
        self.best_params_: Dict = {}
        self.best_model_: Any = None

    def fit(self, train_embeddings: np.array, train_labels: List[str], 
            test_embeddings: np.array, true_labels: List[str]) -> None:
        """
        Train models for all parameter combinations from the grid.

        Args:
            train_embeddings (Any): Training embeddings.
            train_labels (Any): Training labels.
            test_embeddings (Any): Test embeddings.
            true_labels (Any): True labels for test data.
        """
        
        for params in product(*self.param_grid.values()):
            param_dict: Dict[str, Any] = dict(zip(self.param_grid.keys(), params))
            model: Any = self.model_class(**param_dict)
            model.fit(train_embeddings, train_labels)

            start_time: float = time.time()
            predicted_labels: Any = model.predict(test_embeddings) 
            end_time: float = time.time() 

            accuracy: float = accuracy_score(true_labels, predicted_labels)
            f1_weighted: float = f1_score(true_labels, predicted_labels, average='weighted')
            execution_time: float = end_time - start_time

            self.results['params'].append(param_dict.get('k'))
            self.results['accuracy'].append(accuracy)
            self.results['f1_score'].append(f1_weighted)
            self.results['execution_time'].append(execution_time)

            if f1_weighted > self.best_f1_score_:
                self.best_accuracy_score_ = accuracy
                self.best_f1_score_ = f1_weighted
                self.best_params_ = param_dict
                self.best_model_ = model

    def get_results_df(self) -> pd.DataFrame:
        """
        Get the search results as a DataFrame.

        Returns:
            pd.DataFrame: DataFrame with search results.
        """
        
        return pd.DataFrame(self.results)
````
